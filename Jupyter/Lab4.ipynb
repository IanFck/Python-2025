### B.- Imprimir la cantidad de filas duplicadas existentes en el dataset, y si existen duplicados eliminar las filas. Crear un nuevo dataframe que contenga los pokemons sin las filas duplicadas ### 

Aqui usaremos el comando duplicated(), con el podremos ver las filas duplicadas.

```python
df.duplicated()
df
```

Aqui usaremos un if para ver si existen datos duplicados, lo haremos sumando los datos duplicados, si hay mas de 0 se eliminaran con el comando drop_duplicates, y si no se seguira tal cual.

```python
duplicadosdf = df.duplicated().sum()
if duplicadosdf > 0:
    print(f"Se encontraron {duplicadosdf} datos duplicados, se van a proceder a eliminar")
    df.drop_duplicates(inplace=True)
else:
    print("No se encontraron datos duplicados")

```

```python
df_sinduplicados = df.drop_duplicates()
df_sinduplicados
```

### C.- Contar todos los valores nulos existentes en el dataset (Columnas y Filas)e imprimirlos. Luego, eliminar los valores nulos encontrados. Crear un nuevo dataframe, el cual ya no tendrá datos nulos. ###

### F.- Crear los siguientes gráficos utilizando Matplotlib: ###

Gráfico de Barra: Este gráfico debe mostrar el top 10 de Pokemons tipo fuego por su nivel de ataque.

Para la creacion de graficos, seleccionamos las columnas que queremos graficar, en este caso "Att# (Ataque) y "Name" para eso crearemos un nuevo df llamado pokemons_fuego

```python
pokemons_fuego = df_fuego[["Att", "Name",]]
pokemons_fuego
```

Ahora para el grafico, usaremos .plot y luego le daremos (kind="bar) donde en kind, especificamos el tipo de grafico que queremos en esta caso es un grafico de barras "bar". 
Y luego le agregamos el titulo con plt.title, la parte de abajo del grafico con plt.xlabel, y la parte del costado con plt.ylabel.

```python
grafico = pokemons_fuego.head(10)
grafico.set_index("Name", inplace=True)
grafico.plot(kind="bar")
plt.title("Top 10 Pokemons tipo fuego")
plt.xlabel ("Pokemons por nombres")
plt.ylabel("Ataque de pokemones")
```
