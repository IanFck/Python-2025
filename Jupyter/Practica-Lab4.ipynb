```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("Data/ventas.csv", index_col="ID_Venta")
df
```

```python
df["Total venta"] = df["Cantidad"] * df["Precio_Unitario"]
df
```

```python
df.isnull().any()
```

```python
total_ventas = df[["Total venta","Vendedor"]]
total_ventas
```

```python
ventas_por_region = df[["Total venta", "Region"]]
ventas_por_region
```

```python
ventas_por_region
ventas_por_region.plot(kind="bar", color="skyblue", edgecolor="black")#Esto es para poner mas color
plt.title("Ventas totales por region")
plt.xlabel("Region")
plt.ylabel("Total de ventas")
plt.tight_layout()
plt.show()
```
